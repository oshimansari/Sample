```python
# Find the largest of three numbers
```

```python
a = int(input("Enter the first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
```

```python
if a >= b and a >= c:
    print("Largest is:", a)
elif b >= a and b >= c:
    print("Largest is:", b)
else:
    print("Largest is:", c)
```
